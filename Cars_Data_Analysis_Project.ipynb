**Data Cleaning**

```python
# isnull function is used to check the null values present in the dataset
# here false means no null value is present here
cds.isnull()
```

```python
#sum() function used to find all null values count based on the column - in this case
cds.isnull().sum()
```

fill null values of each column with mean value of that column or with a default value

```python
cds['Make'].fillna('Acura',inplace=True)
cds['Model'].fillna('C70 LPT convertible 2dr',inplace=True)
cds['Type'].fillna('Sedan',inplace=True)
cds['Origin'].fillna('Europe',inplace=True)
cds['DriveTrain'].fillna('Front',inplace=True)
cds['MSRP'].fillna('$40,565',inplace=True)
cds['Invoice'].fillna('$40,083',inplace=True)
# fillna() func used to fill the null values with given values in the entire column of the dataset
# Here Make,Model,Type,Origin,DriveTrain,MSRP,Invoice is str so we cannot find mean so, we have given a default value 
# here inplace=True makes the permanent changes of the null values to the given values
```

```python
cds['EngineSize'].fillna(cds['EngineSize'].mean(),inplace=True)
cds['Cylinders'].fillna(cds['Cylinders'].mean(),inplace=True)
cds['Horsepower'].fillna(cds['Horsepower'].mean(),inplace=True)
cds['MPG_City'].fillna(cds['MPG_City'].mean(),inplace=True)
cds['MPG_Highway'].fillna(cds['MPG_Highway'].mean(),inplace=True)
cds['Weight'].fillna(cds['Weight'].mean(),inplace=True)
cds['Wheelbase'].fillna(cds['Wheelbase'].mean(),inplace=True)
cds['Length'].fillna(cds['Length'].mean(),inplace=True)
# this all are int values so we can find mean and replace the null values
```

```python
cds.isnull().sum()
# Now the data has no null values
```
